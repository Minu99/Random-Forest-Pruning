```python
#Import Statements
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
#Data Set Reading
# list for column headers
names = ['Preg', 'Plas', 'Pres', 'Skin', 'Test', 'Mass', 'Pedi', 'Age', 'Class']
df = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv",names=names)
global N,M
N,M=df.shape
df.head()
#To convert into Target and Attributes
X_Data = df.drop("Class", axis=1)
y_Target = df["Class"]
print(N,M)
```

```python

def data_split():    
    #To split the dataset into 3 parts   
    X_train, X_test, y_train, y_test= train_test_split(X_Data, y_Target, test_size=0.4)
    X_train, X_val, y_train, y_val= train_test_split(X_train, y_train, test_size=0.5)
    
    return X_train,X_val,X_test,y_train,y_val,y_test
```

```python
X_train,X_val,X_test,y_train,y_val,y_test=data_split()
print(X_train.shape,X_val.shape,X_test.shape,y_train.shape,y_val.shape,y_test.shape)
```

```python
K=int(M**(1/2))
# Random forest model 
randomforest = RandomForestClassifier(n_estimators=300,max_features=K)
# randomforest.fit(X_train,y_train)
# Prediction of the model
# randomforest_predict = randomforest.predict(X_test)
```

```python
randomforest.fit(X_train,y_train)
print(randomforest.estimators_)
print(K)
```

```python

print(randomforest.estimators_[1])
```

```python
print(randomforest.estimators_[1].predict(X_train))
randomforest.estimators_[1].score(X_train,y_train)
```

```python
#To create a new random forest object and to append only one tree to the forest 
#from existing random foret
randomforestSBS = RandomForestClassifier()
estimatorsSBS=[]
estimatorsSBS.append(randomforest.estimators_[1])
```

```python
#Sbstitution of required parameter in the new random forest object
randomforestSBS.estimators_=estimatorsSBS
randomforestSBS.n_estimators=len(estimatorsSBS)
randomforestSBS.classes_=randomforest.classes_
randomforestSBS.n_classes_=randomforest.n_classes_
randomforestSBS.n_outputs_=randomforest.n_outputs_
```

```python
#Check -If the prediction of the single decision tree and random forest is equal
randomforestSBS.predict(X_train)
randomforestSBS.score(X_train,y_train)
```

```python
#To combine more than one decision tree into a random forest classifier
randomforestCombine = RandomForestClassifier()
randomforestCombine.estimators_=[randomforest.estimators_[1],randomforest.estimators_[4]]
randomforestCombine.n_estimators=len(estimatorsSBS)
randomforestCombine.classes_=randomforest.classes_
randomforestCombine.n_classes_=randomforest.n_classes_
randomforestCombine.n_outputs_=randomforest.n_outputs_
```
